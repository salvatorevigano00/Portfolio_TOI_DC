```python
def gen_codici(prefissi, lunghezza):
    # Caso base: lunghezza 0, restituisci il set di prefissi
    if lunghezza == 0:
        return set(prefissi)
    
    # Altrimenti, crea un set per contenere i suffissi delle parole di codice generate
    suffissi = set()
    
    # Richiama la funzione in modo ricorsivo per ottenere il set di codici di lunghezza 'lunghezza-1'
    codici_precedenti = gen_codici(prefissi, lunghezza - 1)

    # Per ogni parola nel set di prefissi
    for parola in prefissi:
        # Controlla se ci sono suffissi di codici di lunghezza 'lunghezza-1' che possono essere concatenati a 'parola'
        for codice_prec in codici_precedenti:
            if len(parola) > len(codice_prec) and parola.startswith(codice_prec):
                suffissi.add(parola[len(codice_prec):])
        
    # Fa lo stesso controllo anche per tutti i prefissi di codici di lunghezza 'lunghezza-1'
    for codice_prec in codici_precedenti:
        for parola in prefissi:
            if len(codice_prec) > len(parola) and codice_prec.startswith(parola):
                suffissi.add(codice_prec[len(parola):])
        
    return suffissi

# Genera il set di parole di codice infinite a partire da un set di parole di codice iniziale 'set_codici_iniziale'
def gen_codici_infiniti(set_codici_iniziale):
    codici_precedenti = []
    set_codici_infiniti = set()
    lunghezza = 1

    # Continua finché non viene rilevato un ciclo di codici
    while True:
        # Genera il set di codici di lunghezza 'lunghezza'
        set_codici_corrente = gen_codici(set_codici_iniziale, lunghezza)
        print('c_{}'.format(lunghezza), set_codici_corrente)

        # Controlla se il set di codici corrente è già stato generato in precedenza (rilevamento del ciclo)
        if set_codici_corrente in codici_precedenti:
            print('Ciclo rilevato. Algoritmo interrotto.')
            break
        else:
            # Aggiunge il set di codici corrente alla lista dei set di codici generati in precedenza
            codici_precedenti.append(set_codici_corrente)
            # Aggiunge il set corrente al set di codici infiniti
            set_codici_infiniti.update(set_codici_corrente)
            lunghezza += 1
    
    return set_codici_infiniti

# Verifica se gli elementi di 'set_codici' sono decodificabili in modo univoco
def decodifica_univoca(set_codici):
    # Genera il set di codici infiniti
    set_codici_infiniti = gen_codici_infiniti(set_codici)
    # Verifica se l'intersezione tra 'set_codici' e 'set_codici_infiniti' è vuota
    return len(set_codici.intersection(set_codici_infiniti)) == 0

# Stampa se gli elementi di 'set_codici' sono decodificabili in modo univoco o meno
def verifica_decodificabilita(set_codici):
    if decodifica_univoca(set_codici):
        print(set_codici, 'è decodificabile in modo univoco')
    else:
        print(set_codici, 'non è decodificabile in modo univoco')

# Esempi di set di codici da verificare
set_codici = set(['012', '0123', '4', '310', '1024', '2402', '2401', '4013'])
verifica_decodificabilita(set_codici)
print('----------------------------')
set_codici1 = set(['10', '010', '1', '1110'])
verifica_decodificabilita(set_codici1)
print('----------------------------')
set_codici2 = set(['0', '001', '101', '11'])
verifica_decodificabilita(set_codici2)
print('----------------------------')
set_codici3 = set(['0', '2', '3', '011', '104', '341', '11234'])
verifica_decodificabilita(set_codici3)
print('----------------------------')
set_codici4 = set(['01', '10', '001', '100', '000', '111'])
verifica_decodificabilita(set_codici4)
```

```python
# Huffman Coding in python


# stringa = 'BCAADDDCCACACAC'
stringa = 'AAAAAAAAAAEEEEEEEEEEEEEEEIIIIIIIIIIIISSSTTTTPPPPPPPPPPPPPF'

# Definizione della classe dei nodi dell'albero di Huffman
class NodeTree(object):

    def __init__(self, sinistra=None, destra=None):
        self.sinistra = sinistra
        self.destra = destra

    def children(self):
        return (self.sinistra, self.destra)

    def nodes(self):
        return (self.sinistra, self.destra)

    def __str__(self):
        return '%s_%s' % (self.sinistra, self.destra)


# Funzione ricorsiva per la creazione del codice di Huffman
def huffman_code_tree(node, sinistra=True, binString=''):
    # Se il nodo è una foglia, restituisce il suo codice binario
    if type(node) is str:
        return {node: binString}
    # Se il nodo ha figli, richiama la funzione ricorsivamente per i figli sinistro e destro
    (l, r) = node.children()
    d = dict()
    d.update(huffman_code_tree(l, True, binString + '0'))
    d.update(huffman_code_tree(r, False, binString + '1'))
    return d


# Calcolo delle frequenze dei caratteri nella stringa
freq = {}
for c in stringa:
    if c in freq:
        freq[c] += 1
    else:
        freq[c] = 1

# Ordinamento delle frequenze in ordine decrescente
freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)

nodes = freq

while len(nodes) > 1:
    (key1, c1) = nodes[-1]
    (key2, c2) = nodes[-2]
    nodes = nodes[:-2]
    node = NodeTree(key1, key2)
    nodes.append((node, c1 + c2))
    nodes = sorted(nodes, key=lambda x: x[1], reverse=True)

# Generazione del codice di Huffman per ciascun carattere
huffmanCode = huffman_code_tree(nodes[0][0])

caratteri = ""

print(' Carattere | Huffman code ')
print('----------------------')
for (char, frequency) in freq:
    print(' %-9r |%10s' % (char, huffmanCode[char]))
    caratteri += char

# Codifica della stringa originale utilizzando il codice di Huffman
codifica = ""
for lettera in stringa:
    codifica += huffmanCode[lettera]

print("\nCodice codificato\n", codifica)

# Decodifica della stringa codificata utilizzando il dizionario di Huffman
decodifica = ""
codice = ""
dizionario = dict(huffmanCode)

for bit in codifica:
    codice += bit
    for lettera in caratteri:
        if codice == dizionario[lettera]:
            decodifica += lettera
            codice = ""
            
print("\nCodice decodificato\n", decodifica)
```

```python
# Definizione della funzione di codifica Levenshtein
def levenshtein(n):
    if n == 0:
        return "0"
    else:
        C = 1
        code = ""

        binary_str = bin(n)[3:]  # Elimino 0b e il primo 1
        M = len(binary_str)
        while M != 0:
            C += 1
            code = binary_str + code
            binary_str = bin(M)[3:]
            M = len(binary_str)
            
        code = "1" * C + "0" + code

        return code

# Definizione della funzione di decodifica Levenshtein
def levenshtein_dec(n):
    counter = 0
    n_grande = 1
    bit_da_leggere = n_grande

    # Conto gli 1 fino al primo 0
    for bit in n:
        if bit == "1":
            counter += 1
        else:
            break

    n = n[counter + 1:]

    # Itero per recuperare la codifica originale
    for n_grande in range(counter - 2):  # Mi fermo uno prima perché l'ultimo è la codifica
        finestra = "1" + n[:bit_da_leggere]  # Prendo N bit da leggere e aggiungo un 1 all'inizio
        n = n[bit_da_leggere:]  # Tolgo i bit letti
        bit_da_leggere = int(finestra, 2)  # Converto in decimale gli N bit letti

    # Aggiungo un 1 all'inizio e converto in decimale
    decodifica = int(("1" + n), 2)
    return decodifica

# Esecuzione della codifica e decodifica di un numero intero
codifica = levenshtein(23)
decodifica = levenshtein_dec(codifica)

# Stampa dei risultati
print("\nCodice Codificato\n", codifica)
print("\nCodice decodificato\n", decodifica)
```

```python
def lzss_codifica(stringa, finestra, lunghezza_massima):
    posizione_corrente = 0
    risultato = []
    
    while posizione_corrente < len(stringa):
        # Definizione della finestra di ricerca
        finestra_inizio = max(0, posizione_corrente - finestra)
        finestra_fine = posizione_corrente
        finestra_corrente = stringa[finestra_inizio:finestra_fine]

        # Ricerca di una corrispondenza nella finestra
        distanza, lunghezza_match = ricerca_match(stringa, finestra_corrente, posizione_corrente, lunghezza_massima)

        if lunghezza_match > 0:
            # Se è stata trovata una corrispondenza, aggiungo le informazioni corrispondenti alla codifica
            risultato.append((distanza, lunghezza_match))
            posizione_corrente += lunghezza_match
        else:
            # Se non è stata trovata una corrispondenza, aggiungo il carattere alla codifica
            risultato.append((0, stringa[posizione_corrente]))
            posizione_corrente += 1

    return risultato

def ricerca_match(stringa, finestra_corrente, posizione_corrente, lunghezza_massima):
    distanza = 0
    lunghezza_match = 0

    for lunghezza in range(1, min(len(stringa) - posizione_corrente, lunghezza_massima) + 1):
        sottostringa = stringa[posizione_corrente:posizione_corrente + lunghezza]

        if sottostringa in finestra_corrente:
            # Se la sottostringa è presente nella finestra, calcolo la distanza e la lunghezza della corrispondenza
            distanza = posizione_corrente - finestra_corrente.rindex(sottostringa)
            lunghezza_match = lunghezza

    return distanza, lunghezza_match

def lzss_decodifica(codifica):
    decodifica = ""
    finestra = ""
    
    for iter in codifica:
        if iter[0] == 0:
            # Se il primo elemento della tupla è 0, la tupla rappresenta un carattere non compresso
            decodifica += iter[1]
            finestra += iter[1]
        else:
            # Altrimenti, la tupla rappresenta una corrispondenza
            lunghezza = len(finestra)
            
            # Aggiungo la corrispondenza alla decodifica, prendendo la sottostringa dalla finestra
            decodifica += finestra[lunghezza-iter[0] : lunghezza-iter[0]+iter[1]]
            finestra += finestra[lunghezza-iter[0] : lunghezza-iter[0]+iter[1]]
            
            

    return decodifica


#stringa = "abaababaa"
stringa = "abracadabra"
#stringa= "bcababbcbcbaaaabbc"
finestra = 50
lunghezza_massima = 4

codifica = lzss_codifica(stringa, finestra, lunghezza_massima)
print("\nCodice codificato\n", (codifica))

decodifica = lzss_decodifica(codifica)
print("\nCodice decodificato\n", (decodifica))
```
